```python
import pandas as pd
import json
```

```python
df = pd.read_csv("new_train.csv")
```

```python
with open("recipes_non_reasoning.jsonl", "w", encoding="utf-8") as f:
    for _, row in df.iterrows():
        rec = {
            "instruction": row["instruction"],
            "input": row["input"],
            "output": row["output"],
        }
        f.write(json.dumps(rec, ensure_ascii=False) + "\n")
```

```python
reasoning_records = []
for _, row in df.iterrows():
    question = row["instruction"]
    answer = row["output"]
    yemek_adi = str(row["input"])

    reasoning = (
        f"Öncelikle soruyu analiz edelim: {question} "
        f"Burada '{yemek_adi}' yemeğinin adı, malzemeleri, kaç kişilik olduğu ile "
        "hazırlama ve pişirme süreleri istenmektedir. "
        "Bu nedenle cevabımızda önce yemeğin adını, ardından kaç kişilik olduğunu, "
        "sonra hazırlama ve pişirme sürelerini, daha sonra da malzeme listesini ve "
        "yapılış adımlarını açık ve sıralı biçimde yazarız. "
        "Sonuç olarak tarif metninde bu bilgilerin hepsini düzenli bir şekilde sunarız."
    )

    reasoning_records.append(
        {
            "question": question,
            "reasoning": reasoning,
            "answer": answer,
        }
    )

with open("recipes_reasoning.jsonl", "w", encoding="utf-8") as f:
    for rec in reasoning_records:
        f.write(json.dumps(rec, ensure_ascii=False) + "\n")
```

```python
from datasets import load_dataset

# Load reasoning dataset separately
reasoning_dataset = load_dataset(
    "json",
    data_files="recipes_reasoning.jsonl",
    split="train" # default split name for single file
)

# Load non-reasoning dataset separately
non_reasoning_dataset = load_dataset(
    "json",
    data_files="recipes_non_reasoning.jsonl",
    split="train" # default split name for single file
)
```

```python
from datasets import DatasetDict

hf_dataset = DatasetDict({
    "reasoning": reasoning_dataset,
    "non_reasoning": non_reasoning_dataset,
})
```

```python
repo_id = "EndlessSpring/new_trains"

# Push the reasoning dataset as a separate configuration
print(f"Pushing reasoning dataset to {repo_id} with config 'reasoning'")
hf_dataset["reasoning"].push_to_hub(repo_id, config_name="reasoning")

# Push the non-reasoning dataset as a separate configuration
print(f"Pushing non_reasoning dataset to {repo_id} with config 'non_reasoning'")
hf_dataset["non_reasoning"].push_to_hub(repo_id, config_name="non_reasoning")
```
